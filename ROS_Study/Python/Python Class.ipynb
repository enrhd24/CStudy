```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0,6,0.1)
t
```

```python
t.shape
```

```python
y = np.sin(np.pi * t)
y.shape
```

```python
plt.figure(figsize=(12,6))
plt.plot(t,y)
plt.grid()
plt.show()
```

```python
def draw_sin(t, A, f, b):
    y = A * np.sin(np.pi * f * t) + b
    
    plt.figure(figsize=(12,6))
    plt.plot(t,y)
    plt.grid()
    plt.show()
```

```python
t1 = np.arange(0,30,0.01)
draw_sin(t1,1,1,0)
```
